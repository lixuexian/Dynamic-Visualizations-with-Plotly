## Task

The user might be interested in finding the trends of the price for a specific stock. For example, they want to find the trends of price of XYL over time. So, this task is a discover task and the target is to find the trends.

```python
# Find all XYL data
XYL = NYSE_price.loc[NYSE_price["symbol"] == "XYL"]
XYL
```

```python
date = pd.to_datetime(XYL["date"])
lowest_price = XYL["low"]
highest_price = XYL["high"]
open_price = XYL["open"]
close_price = XYL["close"]
```

```python
# Draw four lines
trace_low = go.Scatter(x = date, y = lowest_price, name = "lowest price", mode = "markers + lines")
trace_high = go.Scatter(x = date, y = highest_price, name = "highest price",  mode = "markers + lines")
trace_open = go.Scatter(x = date, y = open_price, name = "open price", mode = "markers + lines")
trace_close = go.Scatter(x = date, y = close_price, name = "close price", mode = "markers + lines")
```

```python
# Subplot and layout
fig_line = tools.make_subplots(rows = 4, cols = 1)
fig_line.append_trace(trace_low, 1, 1)
fig_line.append_trace(trace_high, 2, 1)
fig_line.append_trace(trace_open, 3, 1)
fig_line.append_trace(trace_close, 4, 1)

fig_line["layout"].update(
    title = "Price of XYL over time",
    xaxis1 = dict(title = "Date"),
    yaxis1 = dict(title = "Lowest Price"),
    xaxis2 = dict(title = "Date"),
    yaxis2 = dict(title = "Highest Price"),
    xaxis3 = dict(title = "Date"),
    yaxis3 = dict(title = "Open Price"),
    xaxis4 = dict(title = "Date"),
    yaxis4 = dict(title = "Close Price"),
    autosize = True,
    hovermode = "closest"
)

py.iplot(fig_line, filename = "XYL_chart_1")
```

I create a line chart to visualize the  price of XYL stock over time. Line chart is especially suitable for the target and can easily find the trends of the data. But there is a problem of this plot, there is a large position of overlapping between two date time and viewers need to zoom in to see a clearer picture. 

Since I fix the symbol to "XYL", the key of this chart is date. The date are encoded using the position on the x-axis. The lowest,highest, open and close price over time are encoded using the points & lines as the marks and position & color as the channels. Also I use four sublot to visualize the data to avoid overlapping among those lines.

In this plot we can find that from 2013 to 2017, the price of XYL stock increase. And from 2016 to 2017 the price increase dramatically.

I use closest hovering, zoom and drag as the interactivity. Since there is a large position of overlapping on this plot, viewers can only see the general trend of the price. By using those interactivity, this plot enable viewers to zoom in and drag to see the detail of data. Also the closest hovering enable viewers to see the corresponding x value and y value of the data when mouse move on one data point.

```python
# Draw Candlestick chart
trace = go.Candlestick(
    x = date, 
    low = lowest_price,
    high = highest_price,
    open = open_price,
    close = close_price
)
```

```python
# Layout
layout = go.Layout(
    title = "Lowest and highest price of XYL over time",
    xaxis = dict(title = "Date"),
    yaxis = dict(title = "Price"),
    autosize = True,
    hovermode = "closest"
)
fig_candlestick = dict(data = [trace], layout = layout)
py.iplot(fig_candlestick, filename = "XYL chart 2")
```

I create a candlestick plot to visualize the price of XYL stock over time.In this chart the can show open, close, lowest and highest price for each day in one data point. Also it use color as the channel to show the trend of the data. Red color means the price is decreasing. Green color means the price is increasing.The date time are encoded using the position on the x-axis. The price are encoded using the line as the mark and position & color as the channels.

In this plot there is a gap on the price around Aug 2013 and Nov 2013 which indicate the company may undergoing a hard time.

I use the default interactivity settings and closest hovering. The cloest hovering enable viewers to check the corresponding price of one specific day. The default setting of candlestick plot enable viewers to select a specific time period which helping them better discover the price over time.
